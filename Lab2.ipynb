## 5. Fix Types

changed the votes column to integers
filled each NaN with 0
replaced all commas with space for clear data 

```python

df['votes'] = df['votes'].fillna(0)
df['votes'] = df['votes'].replace(',','', regex=True)
df['votes'] = df['votes'].astype(int)
```

got rid of the annoying * before the 9.0
made each Nan a 0
made the data entrys all float objects to do statistics to later

```python
df['rating'] = df['rating'].str.replace('*','', regex=True)
df['rating'] = df['rating'].replace('Nan',0)
df['rating'] = df['rating'].astype(float)
```
